```python
import json
import pandas as pd

# Load JSON data
df_stream0 = pd.read_json('StreamingHistory0.json')
df_stream1 = pd.read_json('StreamingHistory1.json')

# Merge streaming dataframes
df_stream = pd.concat([df_stream0, df_stream1], ignore_index=True)

# Convert timestamp to datetime
df_stream['endTime'] = pd.to_datetime(df_stream['endTime'])
```

```python
top_tracks = df_stream['trackName'].value_counts().head(10)
print("Top Tracks:")
print(top_tracks)

# Top artists
top_artists = df_stream['artistName'].value_counts().head(5)
print("Top Artists:")
print(top_artists)
```

```python
# Convert timestamp to datetime
df_stream['endTime'] = pd.to_datetime(df_stream['endTime'])

# Listening habits by hour of the day
df_stream['hour_of_day'] = df_stream['endTime'].dt.hour
hour_order = list(range(24))
hour_of_day_counts = df_stream['hour_of_day'].value_counts().reindex(hour_order, fill_value=0)
print("Listening Patterns by Hour of the Day:")
print(hour_of_day_counts)

# Calculate total listening time
df_stream['hours_played'] = df_stream['msPlayed'] / (1000 * 60 * 60)

# Listening time by hour of the day
listening_time_by_hour = df_stream.groupby('hour_of_day')['hours_played'].sum().reindex(hour_order, fill_value=0)
print("Total Listening Time by Hour of the Day (in hours):")
for hour, hours in listening_time_by_hour.items():
    print(f"{hour}:00 - {hour}:59: {hours:.2f} hours")
```

```python
# Listening habits by day of the week
df_stream['day_of_week'] = df_stream['endTime'].dt.day_name()
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
day_of_week_counts = df_stream['day_of_week'].value_counts().reindex(day_order)
print("Listening Patterns by Day of the Week:")
for day, count in day_of_week_counts.items():
    print(f"{day}: {count} tracks played")

# Listening time by day of the week
listening_time_by_day = df_stream.groupby('day_of_week')['hours_played'].sum().reindex(day_order)
print("Total Listening Time by Day of the Week (in hours):")
for day, hours in listening_time_by_day.items():
    print(f"{day}: {hours:.2f} hours")
```

```python
# Aggregate top tracks and artists
top_tracks = df_stream['trackName'].value_counts().head(10)
top_artists = df_stream['artistName'].value_counts().head(10)

# Aggregate most active listening hours and days
most_active_hours = df_stream['hour_of_day'].value_counts().head(5)
most_active_days = df_stream['day_of_week'].value_counts().head(5)

print("Top Tracks:")
print(top_tracks)
print("\nTop Artists:")
print(top_artists)
print("\nMost Active Listening Hours:")
print(most_active_hours)
print("\nMost Active Listening Days:")
print(most_active_days)
```

```python
# Calculate unique counts and total minutes
unique_artists = df_stream['artistName'].nunique()
unique_tracks = df_stream['trackName'].nunique()
total_minutes = df_stream['msPlayed'].sum() / 60000

print(f"Count of Unique Artists Listened: {unique_artists}")
print(f"Count of Unique Songs Listened: {unique_tracks}")
print(f"Total Minutes Listened: {total_minutes:.2f}")

# Extract month from endTime
df_stream['month'] = df_stream['endTime'].dt.to_period('M')

# Aggregate listening habits by month
listening_by_month = df_stream['month'].value_counts().sort_index()

print("Listening Habits by Month:")
print(listening_by_month)
```

```python
def get_genres_for_tracks_and_artists(top_tracks, top_artists, sp):
    track_genres = {}
    artist_genres = {}

    # Fetch genres for top tracks
    for track in top_tracks.index:
        results = sp.search(q=f"track:{track}", type='track')
        if results['tracks']['items']:
            track_id = results['tracks']['items'][0]['id']
            artist_id = results['tracks']['items'][0]['artists'][0]['id']
            artist = sp.artist(artist_id)
            track_genres[track] = artist['genres']

    # Fetch genres for top artists
    for artist in top_artists.index:
        results = sp.search(q=f"artist:{artist}", type='artist')
        if results['artists']['items']:
            artist_id = results['artists']['items'][0]['id']
            artist = sp.artist(artist_id)
            artist_genres[artist['name']] = artist['genres']

    return track_genres, artist_genres

# Get genres for top tracks and artists
track_genres, artist_genres = get_genres_for_tracks_and_artists(top_tracks, top_artists, sp)

print("Track Genres:")
for track, genres in track_genres.items():
    print(f"{track}: {genres}")

print("\nArtist Genres:")
for artist, genres in artist_genres.items():
    print(f"{artist}: {genres}")
```

```python
def get_recommendations(top_tracks, top_artists, sp):
    recommendations = []
    for track in top_tracks.index:
        results = sp.search(q=f"track:{track}", type='track')
        if results['tracks']['items']:
            track_id = results['tracks']['items'][0]['id']
            recs = sp.recommendations(seed_tracks=[track_id], limit=5)
            for rec in recs['tracks']:
                recommendations.append(rec['name'])
    for artist in top_artists.index:
        results = sp.search(q=f"artist:{artist}", type='artist')
        if results['artists']['items']:
            artist_id = results['artists']['items'][0]['id']
            recs = sp.recommendations(seed_artists=[artist_id], limit=5)
            for rec in recs['tracks']:
                recommendations.append(rec['name'])
    return recommendations

# Get recommendations
recommendations = get_recommendations(top_tracks, top_artists, sp)
unique_recommendations = list(set(recommendations))  # Remove duplicates

# Get a set of already listened track names
listened_tracks = set(df_stream['trackName'])

# Separate recommendations into new and already listened tracks
new_recommendations = [rec for rec in unique_recommendations if rec not in listened_tracks]
already_listened_recommendations = [rec for rec in unique_recommendations if rec in listened_tracks]

print("New Recommendations (tracks you haven't listened to yet):")
for rec in new_recommendations:
    print(rec)

print("\nAlready Listened Recommendations (tracks you have listened to):")
for rec in already_listened_recommendations:
    print(rec)
```
